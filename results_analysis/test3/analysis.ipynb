```python
import pandas as pd
import matplotlib.pyplot as plt

# 读取数据
vllm_df = pd.read_csv('throughput_vllm.csv')
booking_df = pd.read_csv('throughput_booking_limit_400.csv')
booking_df2 = pd.read_csv('throughput_booking_limit_1000.csv')
booking_df3 = pd.read_csv('throughput_booking_limit_44.csv')
booking_df4 = pd.read_csv('throughput_booking_limit_90.csv')
sarathi_df = pd.read_csv('throughput_sarathi.csv')

# 设置全局字体
plt.rcParams['font.family'] = 'serif'
plt.rcParams['font.size'] = 12

# 设置时间范围
T0 = 50  # 设定起始时间
T = 180  # 设定结束时间

# 计算 T0 时刻的 throughput
def get_T0_throughput(df, T0):
    """ 获取 T0 时刻的 throughput，如果没有精确的 T0，使用最近的 """
    df_T0 = df[df['Time (sec)'] <= T0]
    if not df_T0.empty:
        return df_T0.iloc[-1]['throughput']  # 取 T0 时刻或最接近 T0 的值
    return 0  # 如果没有数据，返回 0

# 计算相对 T0 的 throughput
def adjust_throughput(df, T0):
    """ 计算 t 时刻的 throughput 减去 T0 时刻的 throughput """
    T0_throughput = get_T0_throughput(df, T0)
    df = df[df['Time (sec)'] >= T0].copy()  # 过滤 T0 之前的数据
    df['Time (sec)'] -= T0  # 让 T0 作为新的起点 (t=0)
    df['throughput'] -= T0_throughput  # 计算 throughput 相对增长量
    return df

# 处理数据
vllm_subset = adjust_throughput(vllm_df, T0)
sarathi_subset = adjust_throughput(sarathi_df, T0)
booking_subset = adjust_throughput(booking_df, T0)
booking2_subset = adjust_throughput(booking_df2, T0)
booking3_subset = adjust_throughput(booking_df3, T0)
booking4_subset = adjust_throughput(booking_df4, T0)

# 画图
plt.figure(figsize=(10, 6))

plt.plot(vllm_subset['Time (sec)'], vllm_subset['throughput'], 
         color='royalblue', linestyle='-', linewidth=1, alpha=0.8, label='vLLM', zorder=3)

plt.plot(sarathi_subset['Time (sec)'], sarathi_subset['throughput'], 
         color='crimson', linestyle='-', linewidth=1, alpha=0.9, label='Sarathi', zorder=2)

plt.plot(booking_subset['Time (sec)'], booking_subset['throughput'], 
         color='darkred', linestyle='--', linewidth=1, alpha=0.6, label='Booking Limit, total_limit=400', zorder=1)

plt.plot(booking2_subset['Time (sec)'], booking2_subset['throughput'],
         color='green', linestyle='-.', linewidth=1, alpha=0.6, label='Booking Limit, total_limit=1000', zorder=1)

plt.plot(booking3_subset['Time (sec)'], booking3_subset['throughput'],
         color='orange', linestyle='-', linewidth=1, alpha=0.6, label='Booking Limit, total_limit=44', zorder=1)

plt.plot(booking4_subset['Time (sec)'], booking4_subset['throughput'],
         color='purple', linestyle='-', linewidth=1, alpha=0.6, label='Booking Limit, total_limit=90', zorder=1)

# 添加标签和图例
plt.xlabel('Time since T0 (sec)', fontsize=14)
plt.ylabel('Throughput Increase from T0', fontsize=14)
plt.legend(frameon=False, fontsize=12)

# 设置坐标轴格式
plt.grid(True, linestyle=':', alpha=0.7)
plt.ticklabel_format(axis='x', style='plain')  # 禁用科学计数法
plt.tight_layout()

# 显示图形
plt.show()
plt.savefig('throughput_comparison_T0.png', dpi=300, bbox_inches='tight')
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 读取数据
vllm_df = pd.read_csv('throughput_vllm.csv')
booking_df = pd.read_csv('throughput_booking_limit_400.csv')
booking_df2 = pd.read_csv('throughput_booking_limit_1000.csv')
booking_df3 = pd.read_csv('throughput_booking_limit_44.csv')
booking_df4 = pd.read_csv('throughput_booking_limit_90.csv')
sarathi_df = pd.read_csv('throughput_sarathi.csv')

# 设置时间范围
T0 = 110  # 设定起始时间
T = 170  # 设定结束时间

# 计算 T0 时刻的 throughput
def get_T0_throughput(df, T0):
    """ 获取 T0 时刻的 throughput，如果没有精确的 T0，使用最近的 """
    df_T0 = df[df['Time (sec)'] <= T0]
    if not df_T0.empty:
        return df_T0.iloc[-1]['throughput']  # 取 T0 时刻或最接近 T0 的值
    return 0  # 如果没有数据，返回 0

# 计算 (t 时刻的 throughput - T0 时刻的 throughput) / (t - T0)
def compute_normalized_throughput(df, T0):
    """ 计算相对 T0 的 throughput 速率 """
    T0_throughput = get_T0_throughput(df, T0)
    df = df[df['Time (sec)'] >= T0].copy()  # 过滤 T0 之前的数据
    df['Time (sec)'] -= T0  # 让 T0 作为新的起点 (t=0)
    df['throughput_per_time'] = (df['throughput'] - T0_throughput) / df['Time (sec)']
    return df

# 处理数据
vllm_subset = compute_normalized_throughput(vllm_df, T0)
sarathi_subset = compute_normalized_throughput(sarathi_df, T0)
booking_subset = compute_normalized_throughput(booking_df, T0)
booking2_subset = compute_normalized_throughput(booking_df2, T0)
booking3_subset = compute_normalized_throughput(booking_df3, T0)
booking4_subset = compute_normalized_throughput(booking_df4, T0)

# 画图
# 画图
plt.figure(figsize=(10, 6))

plt.plot(vllm_subset[vllm_subset['Time (sec)'] >= 1]['Time (sec)'], 
         vllm_subset[vllm_subset['Time (sec)'] >= 1]['throughput_per_time'], 
         color='royalblue', linestyle='-', linewidth=1, alpha=0.8, label='vLLM')

plt.plot(sarathi_subset[sarathi_subset['Time (sec)'] >= 1]['Time (sec)'], 
         sarathi_subset[sarathi_subset['Time (sec)'] >= 1]['throughput_per_time'], 
         color='crimson', linestyle='-', linewidth=1, alpha=0.9, label='Sarathi')

plt.plot(booking_subset[booking_subset['Time (sec)'] >= 1]['Time (sec)'], 
         booking_subset[booking_subset['Time (sec)'] >= 1]['throughput_per_time'], 
         color='darkred', linestyle='--', marker='x', markersize=2, linewidth=1, alpha=0.6, label='Booking Limit, total_limit=400')

plt.plot(booking2_subset[booking2_subset['Time (sec)'] >= 1]['Time (sec)'], 
         booking2_subset[booking2_subset['Time (sec)'] >= 1]['throughput_per_time'], 
         color='green', linestyle='-.', marker='x', markersize=2, linewidth=1, alpha=0.6, label='Booking Limit, total_limit=1000')

plt.plot(booking4_subset[booking4_subset['Time (sec)'] >= 1]['Time (sec)'], 
         booking4_subset[booking4_subset['Time (sec)'] >= 1]['throughput_per_time'], 
         color='purple', linestyle='-', linewidth=1, alpha=0.6, label='Booking Limit, total_limit=90')

# 设置标签和图例
plt.xlabel('Time since T0 (sec)', fontsize=14)
plt.ylabel('Normalized Throughput (ΔThroughput/ΔTime)', fontsize=14)
plt.legend(frameon=False, fontsize=12)
plt.grid(True, linestyle=':', alpha=0.7)
plt.ticklabel_format(axis='x', style='plain')  # 禁用科学计数法
plt.tight_layout()

# 显示图形
plt.show()
```
